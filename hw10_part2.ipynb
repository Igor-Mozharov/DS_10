```python
import numpy as np
from tensorflow import keras
from keras.datasets import fashion_mnist
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Dropout
from keras import losses, metrics, optimizers
from keras.utils import to_categorical
import matplotlib.pyplot as plt
from keras.applications import VGG16
```

```python
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
x_train, x_test = np.array(x_train, np.float32), np.array(x_test, np.float32)
x_train, x_test = x_train / 255., x_test / 255.
# x_train, x_test = np.expand_dims(x_train, axis=3), np.expand_dims(x_test, axis=3)
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
x_train = np.dstack([x_train] * 3).reshape(-1, 28, 28, 3)
x_test = np.dstack([x_test] * 3).reshape(-1, 28, 28, 3)
x_train = keras.preprocessing.image.smart_resize(x_train, (48, 48))
x_test = keras.preprocessing.image.smart_resize(x_test, (48, 48))
```

```python
vgg_model = VGG16(weights='imagenet', include_top=False, input_shape=(48, 48, 3))
vgg_model.trainable = True
set_trainable = False
for layer in vgg_model.layers:
    if layer.name == 'block5_conv1':
        set_trainable = True
    if set_trainable:
        layer.trainable = True
    else:
        layer.trainable = False
```

```python
model = Sequential()
model.add(vgg_model)
model.add(Flatten())
model.add(Dense(1280, activation='relu'))
model.add(Dense(650, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation='softmax'))
model.compile(optimizer=optimizers.Nadam(learning_rate=0.001), metrics=[metrics.accuracy],
loss=losses.binary_crossentropy)
model.summary()
```

```python
res = model.fit(x_train, y_train, batch_size=50, epochs=70, validation_split=0.35)
```

```python
htr = res.history
loss = htr['loss']
val_loss = htr['val_loss']
accuracy = htr['accuracy']
val_accuracy = htr['val_accuracy']
epochs = range(1, len(accuracy) + 1)
```

```python
plt.plot(epochs, loss, label='loss')
plt.plot(epochs, val_loss, label='val_loss')
plt.title('val_loss + loss')
plt.legend()
plt.show()
```

```python
plt.plot(epochs, accuracy, label='accuracy')
plt.plot(epochs, val_accuracy, label='val_accuracy')
plt.title('val_accuracy + accuracy')
plt.legend()
plt.show()
```
